## Часть I. Приоритизация гипотез

### Постановка задачи

Провести приоритизацию гипотез, подготовленных совместно с отделом маркетинга интренет-магазина.

```python
# Импорт библионтек для обеих частей задания
import pandas as pd
pd.options.display.max_colwidth = 130

import datetime as dt

import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters

import warnings
# конвертеры, которые позволяют использовать типы pandas в matplotlib  
register_matplotlib_converters()

import scipy.stats as stats
```

```python
data = pd.read_csv('/datasets/hypothesis.csv')
display(data)
```

**Описание содержания столбцов таблицы с данными**

Hypothesis — краткое описание гипотезы;

Reach — охват пользователей по 10-балльной шкале;

Impact — влияние на пользователей по 10-балльной шкале;

Confidence — уверенность в гипотезе по 10-балльной шкале;

Efforts — затраты ресурсов на проверку гипотезы по 10-балльной шкале.

### Применение фреймворка ICE

```python
data['ICE'] = (data['Impact'] * data['Confidence'] / data['Efforts']).round(2)
data_ICE = data[['Hypothesis', 'ICE']].sort_values(by='ICE', ascending=False)
display(data_ICE)
print('Наиболее приоритетная гипотеза:', data_ICE.iloc[0, 0], sep='\n')
```

### Применение фреймворка RICE

```python
data['RICE'] = (data['Reach'] * data['Impact'] * data['Confidence'] / data['Efforts']).round(2)
data_RICE = data[['Hypothesis', 'RICE']].sort_values(by='RICE', ascending=False)
display(data_RICE)
```

```python
print('Наиболее приоритетная гипотеза:', data_RICE.iloc[0, 0], sep='\n')
```

### Итоги приоритизации

1. По итогам применения фреймворка ICE наиболее приоритетной оказалась гипотеза "Запустить акцию, дающую скидку на товар в день рождения". Однако опережение по индексу ICE от следующей по приоритету гипотезы было не столь значительным (16,2 против 13,3).

2. По итогам применения фреймворка RICE наиболее приоритетной оказалась гипотеза "Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок", которая по методу ICE находилась только на третьей позиции. Причем отрыв этой гипотезы от следующей по приоритету стал более чем существенным - двукратным (112 против 56).

3. Лидировавшая по методу ICE гипотеза "Запустить акцию, дающую скидку на товар в день рождения" опустилась по методу RICE на 5-ю позицию.

4. Причина подъема вверх и безусловного лидерства гипотезы "Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок" по фреймворку RICE заключается в том, что охват пользователей этой гипотезы максимальный (10), что наряду с достаточно высокими баллами Impact и Confidence, а также средним уровнем Reach позволило ей выйти на первую позицию с большим отрывом.

5. Лидировавшая по ICE гипотеза "Запустить акцию, дающую скидку на товар в день рождения" столь существенно понизилась в приоритете по фреймворку RICE благодаря минимальному уровню охвата пользователей (1).

6. К применению рекомендуются результаты приоритизации по фрейморку RICE, где безусловный приоритет получила гипотеза "Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок".

## Часть II. Анализ A/B-теста

### Исходные данные и их подготовка
Имеются результаты A/B-теста, представленные в двух файлах.

```python
orders = pd.read_csv('/datasets/orders.csv')
visitors = pd.read_csv('/datasets/visitors.csv')
display(orders, visitors)
```

**Описание содержания столбцов таблицы с данными**

**Заказы (orders)**

transactionId — идентификатор заказа;

visitorId — идентификатор пользователя, совершившего заказ;

date — дата, когда был совершён заказ;

revenue — выручка заказа;

group — группа A/B-теста, в которую попал заказ.

**Посетители (visitors)**

date — дата;

group — группа A/B-теста;

visitors — количество пользователей в указанную дату в указанной группе A/B-теста/

```python
orders.info()
```

```python
visitors.info()
```

```python
first_day = orders['date'].min()
last_day = orders['date'].max()
print(f'Период анализа: {first_day} - {last_day}')
```

Данные понятны, полны, не содержат пропусков.

Однако требуется обработать формат даты в обеих таблицах - для дальнейших вычислений привести к datetime64.

Также переименуем два столбца в таблице orders - для краткости.

```python
orders['date'] = orders['date'].map(lambda x: dt.datetime.strptime(x, '%Y-%m-%d'))
visitors['date'] = visitors['date'].map(lambda x: dt.datetime.strptime(x, '%Y-%m-%d'))

orders.rename(columns={'transactionId':'orderId', 'visitorId':'userId'}, inplace=True)
```

**Группы А и В. Проблемы выборок**

```python
count_A = orders[orders['group'] == 'A']['userId'].nunique()
count_B = orders[orders['group'] == 'B']['userId'].nunique()
count_AB = count_A + count_B
print (f'Пользователей в группе А: {count_A}, {count_A/count_AB:.1%}\
    \nПользователей в группе B: {count_B}, {count_B/count_AB:.1%}\
    \nПользователей в группах А и B: {count_AB}')
```

```python
nunique_users = orders['userId'].nunique()
print(f'Число пользователей, попавших в группы А и В одновременно: {count_AB - nunique_users}, {1 - nunique_users/count_AB:.1%}')
```

Обнаружилось следущее:
1. Группы не равны по количеству (разница 7,6%) - это не столь критично.
2. 5,3% пользователей попали в обе группы - очевидно, ошибка при формировании групп.

В этой ситуации мы можем исключить тех, кто "сидит на двух стульях", но тогда мы не сможем адекватным образом стыковаться с таблицей посетителей, где пользователи тех же групп А и В уже сгруппированы по датам.

С другой стороны, 5,3% пользователей сглаживают различия в группах А и В, нивелируют их различия. То есть, при выевлении различий можно предположить, что на самом деле они сильнее почти на 12%, чем будет видно по итогам анализа. Это существенный фактор.

В итоге, учитывая существенную роль искажений от задвоенности в заказах, из двух зол выбираем меньшее: очищение данных orders от "двойных агентов". Причина: лучше иметь пониженные примерно на 5% по сравнению с фактом абсолютные показатели конверсии в обиих группах, чем пониженную примерно на 12% степень расхождения между группами, оценка которой, собственно и является предметом анализа.

Формирование списка пользователей, имеющих двойную "маркировку" (А и В) и очистка данных от таких пользователей:

```python
users_count = orders.groupby('userId').agg({'group':'nunique'}).reset_index()
double_users = list(users_count[users_count['group'] == 2]['userId'])
print('Число пользователей в списке "двойных агентов":', len(double_users))
```

```python
orders = orders.query('userId not in @double_users')
orders.info()
```

Данные по заказам очищены от "двойных агентов".

### Кумулятивная выручка по группам

```python
datesGroups = orders[['date', 'group']].drop_duplicates()

ordersAggregated = datesGroups.apply(
    lambda x: orders[
        np.logical_and(
            orders['date'] <= x['date'], orders['group'] == x['group']
        )
    ].agg(
        {
            'date': 'max',
            'group': 'max',
            'orderId': 'nunique',
            'userId': 'nunique',
            'revenue': 'sum',
        }
    ),
    axis=1,
).sort_values(by=['date', 'group'])

visitorsAggregated = datesGroups.apply(
    lambda x: visitors[
        np.logical_and(
            visitors['date'] <= x['date'], visitors['group'] == x['group']
        )
    ].agg({'date': 'max', 'group': 'max', 'visitors': 'sum'}),
    axis=1,
).sort_values(by=['date', 'group'])

cumulativeData = ordersAggregated.merge(
    visitorsAggregated, left_on=['date', 'group'], right_on=['date', 'group']
)
cumulativeData.columns = [
    'date',
    'group',
    'orders',
    'buyers',
    'revenue',
    'visitors',
]

cumulativeRevenueA = cumulativeData[cumulativeData['group']=='A'][['date','revenue', 'orders']]
cumulativeRevenueB = cumulativeData[cumulativeData['group']=='B'][['date','revenue', 'orders']]

plt.plot(cumulativeRevenueA['date'], cumulativeRevenueA['revenue'], label='A')
plt.plot(cumulativeRevenueB['date'], cumulativeRevenueB['revenue'], label='B')

plt.legend();
```

**Выводы** 

1. До середины анализируемого периода (августа 2019 года) группа В несколько опережала группу А, однако затем последовал скачок результатов группы В вверх, и ее отрыв стал очень существенным, но стабильным.
2. Скачкообразный рост кумулятивного результата группы В может быть связан с экстремально крупными заказами, сделанными в этот период. Эту гипотезу надо будет проверить.
3. Отдельные не столь крупные, но всё же выбивающиеся из среднестатистического размера заказы могли стать причиной менее значительных изломов обеих кривых роста. Этот предположение также следует проверить.

### Кумулятивный средний чек по группам

```python
plt.plot(cumulativeRevenueA['date'], cumulativeRevenueA['revenue']/cumulativeRevenueA['orders'], label='A')
plt.plot(cumulativeRevenueB['date'], cumulativeRevenueB['revenue']/cumulativeRevenueB['orders'], label='B')
plt.legend();
```

**Выводы** 

1. Кумулятивный средний чек по группам характеризуется значительными колебаниями.
2. Резкий скачок кумулятивного среднего чека в группе В в середине периода и его последующее понижение также свидетельствует в пользу предположения об аномально крупных покупках в этой группе, состоявшихся в середине анализируемого месяца.
3. На данном этипе делать итоговые заключения преждевременно.

### Относительное изменение кумулятивного среднего чека группы B к группе A

```python
mergedCumulativeRevenue = cumulativeRevenueA.merge(cumulativeRevenueB,
                                                   left_on='date', right_on='date', how='left', suffixes=['A', 'B'])

# cтроим отношение средних чеков
plt.plot(mergedCumulativeRevenue['date'], (mergedCumulativeRevenue['revenueB']/mergedCumulativeRevenue['ordersB'])\
         /(mergedCumulativeRevenue['revenueA']/mergedCumulativeRevenue['ordersA'])-1)

# добавляем ось X на нуле
plt.axhline(y=0, color='black', linestyle='--');
```

**Выводы** 

1. Относительное изменение кумулятивного среднего чека группы B к группе A проиллюстрироввло те же перепады во "взаимоотношениях" групп А и В, которые просматривались и предыдущем графике кумулятивных средних чеков.
2. Однако на данном графике хорошо видно, что скачок в группе В (предположительно, обусловленный аномально крупными покупками) и последующее проседание графика группы В относительно А примерно на половину этого скачка свидетельствуют в пользу предположения об аномально крупных покупках в группе В в середине месяца.
3. Значительные колебания графика в начале месяца могут быть следствием влияния случайных факторов в обеих группах, которые  сказывались в начальных период кумуляции результатов обеих групп.

### Кумулятивная конверсия по группам

```python
cumulativeData['conversion'] = cumulativeData['orders']/cumulativeData['visitors']
cumulativeDataA = cumulativeData[cumulativeData['group']=='A']
cumulativeDataB = cumulativeData[cumulativeData['group']=='B']

plt.plot(cumulativeDataA['date'], cumulativeDataA['conversion'], label='A')
plt.plot(cumulativeDataB['date'], cumulativeDataB['conversion'], label='B')
plt.legend()

# Масштаб осей графика задается вручную:
plt.axis([dt.datetime(2019, 8, 1), dt.datetime(2019, 8, 31), 0, 0.05]);
```

**Выводы** 

1. К концу первой недели анализируемого месяца отстававшая группа В вышла вперед.
2. После первой декады кумулятивные конверсии обеих групп стабилизируются и их соотношение держится до конца анализируемого периода.

### Статистическая значимость различий в конверсии между группами по «сырым» данным

```python
visitors
```

```python
orders
```

**Гипотеза Н0**: различий в конверсии между группами А и В нет.

**Гипотеза Н1**: уровни конверсии в группах А и В различаются.

Используем тест Уилкоксона-Манна-Уитни из библиотеки scipy.stats.

Критерием достоверности (максимальный риск ошибки I рода) будем считать 5%.

```python
visitorsADaily = visitors[visitors['group'] == 'A'][['date', 'visitors']]
visitorsADaily.columns = ['date', 'visitorsPerDateA']

visitorsACummulative = visitorsADaily.apply(
    lambda x: visitorsADaily[visitorsADaily['date'] <= x['date']].agg(
        {'date': 'max', 'visitorsPerDateA': 'sum'}
    ),
    axis=1,
)
visitorsACummulative.columns = ['date', 'visitorsCummulativeA']

visitorsBDaily = visitors[visitors['group'] == 'B'][['date', 'visitors']]
visitorsBDaily.columns = ['date', 'visitorsPerDateB']

visitorsBCummulative = visitorsBDaily.apply(
    lambda x: visitorsBDaily[visitorsBDaily['date'] <= x['date']].agg(
        {'date': 'max', 'visitorsPerDateB': 'sum'}
    ),
    axis=1,
)
visitorsBCummulative.columns = ['date', 'visitorsCummulativeB']

ordersADaily = (
    orders[orders['group'] == 'A'][['date', 'orderId', 'userId', 'revenue']]
    .groupby('date', as_index=False)
    .agg({'orderId': pd.Series.nunique, 'revenue': 'sum'})
)
ordersADaily.columns = ['date', 'ordersPerDateA', 'revenuePerDateA']

ordersACummulative = ordersADaily.apply(
    lambda x: ordersADaily[ordersADaily['date'] <= x['date']].agg(
        {'date': 'max', 'ordersPerDateA': 'sum', 'revenuePerDateA': 'sum'}
    ),
    axis=1,
).sort_values(by=['date'])
ordersACummulative.columns = [
    'date',
    'ordersCummulativeA',
    'revenueCummulativeA',
]

ordersBDaily = (
    orders[orders['group'] == 'B'][['date', 'orderId', 'userId', 'revenue']]
    .groupby('date', as_index=False)
    .agg({'orderId': pd.Series.nunique, 'revenue': 'sum'})
)
ordersBDaily.columns = ['date', 'ordersPerDateB', 'revenuePerDateB']

ordersBCummulative = ordersBDaily.apply(
    lambda x: ordersBDaily[ordersBDaily['date'] <= x['date']].agg(
        {'date': 'max', 'ordersPerDateB': 'sum', 'revenuePerDateB': 'sum'}
    ),
    axis=1,
).sort_values(by=['date'])
ordersBCummulative.columns = [
    'date',
    'ordersCummulativeB',
    'revenueCummulativeB',
]

data = (
    ordersADaily.merge(
        ordersBDaily, left_on='date', right_on='date', how='left'
    )
    .merge(ordersACummulative, left_on='date', right_on='date', how='left')
    .merge(ordersBCummulative, left_on='date', right_on='date', how='left')
    .merge(visitorsADaily, left_on='date', right_on='date', how='left')
    .merge(visitorsBDaily, left_on='date', right_on='date', how='left')
    .merge(visitorsACummulative, left_on='date', right_on='date', how='left')
    .merge(visitorsBCummulative, left_on='date', right_on='date', how='left')
)

print(data.head(5))

ordersByUsersA = (
    orders[orders['group'] == 'A']
    .groupby('userId', as_index=False)
    .agg({'orderId': pd.Series.nunique})
)
ordersByUsersA.columns = ['userId', 'orders']

ordersByUsersB = (
    orders[orders['group'] == 'B']
    .groupby('userId', as_index=False)
    .agg({'orderId': pd.Series.nunique})
)
ordersByUsersB.columns = ['userId', 'orders']

sampleA = pd.concat([ordersByUsersA['orders'],pd.Series(0, index=np.arange(data['visitorsPerDateA'].sum() - len(ordersByUsersA['orders'])), name='orders')],axis=0)

sampleB = pd.concat([ordersByUsersB['orders'],pd.Series(0, index=np.arange(data['visitorsPerDateB'].sum() - len(ordersByUsersB['orders'])), name='orders')],axis=0) 

ordersByUsersA = (
    orders[orders['group'] == 'A']
    .groupby('userId', as_index=False)
    .agg({'orderId': pd.Series.nunique})
)
ordersByUsersA.columns = ['userId', 'orders']

ordersByUsersB = (
    orders[orders['group'] == 'B']
    .groupby('userId', as_index=False)
    .agg({'orderId': pd.Series.nunique})
)
ordersByUsersB.columns = ['userId', 'orders']

sampleA = pd.concat(
    [
        ordersByUsersA['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitorsPerDateA'].sum() - len(ordersByUsersA['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)

sampleB = pd.concat(
    [
        ordersByUsersB['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitorsPerDateB'].sum() - len(ordersByUsersB['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)

print('p-valie =', "{0:.3f}".format(stats.mannwhitneyu(sampleA, sampleB)[1]))
print('Относительное различие в конверсии:', "{0:.3f}".format(sampleB.mean() / sampleA.mean() - 1))
```

**Выводы** 

1. р-критерий (1,1%) значительно меньше установленного порогового значения в 5%. В результате у нас есть основания отвергнуть нулевую гипотезу и признать, что конверсии групп А и В, скорее всего, различаются.
2. Относительное различие конверсии (у группы В она выше, чем у группы А) на уровне 16% является значимым для бизнеса.
3. Однако этот расчет сделан на "сырых" данных, включающих в себя выбросы.

### Статистическая значимость различий в конверсии между группами по «очищенным» данным

Произведем очистку данных по ранее выбранным параметрам limit_orders и limit_value.
Для начала определим аномальных пользователей

```python
usersWithManyOrders = pd.concat(
    [
        ordersByUsersA[ordersByUsersA['orders'] > limit_orders]['userId'],
        ordersByUsersB[ordersByUsersB['orders'] > limit_orders]['userId'],
    ],
    axis=0,
)
usersWithExpensiveOrders = orders[orders['revenue'] > limit_value]['userId']
abnormalUsers = (
    pd.concat([usersWithManyOrders, usersWithExpensiveOrders], axis=0)
    .drop_duplicates()
    .sort_values()
)
print(abnormalUsers.head(5))
print(f'Число аномальных пользователей: {abnormalUsers.count()}')
```

Далее произведем расчет, в котором исключены аномальные пользователи (статистические выбросы).

**Гипотеза Н0**: различий в конверсии между группами А и В нет.

**Гипотеза Н1**: уровни конверсии в группах А и В различаются.

Используем тест Уилкоксона-Манна-Уитни из библиотеки scipy.stats.

Критерием достоверности (максимальный риск ошибки I рода) будем считать 5%.

```python
sampleAFiltered = pd.concat(
    [
        ordersByUsersA[
            np.logical_not(ordersByUsersA['userId'].isin(abnormalUsers))
        ]['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitorsPerDateA'].sum() - len(ordersByUsersA['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)

sampleBFiltered = pd.concat(
    [
        ordersByUsersB[
            np.logical_not(ordersByUsersB['userId'].isin(abnormalUsers))
        ]['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitorsPerDateB'].sum() - len(ordersByUsersB['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)
print(sampleBFiltered.value_counts())
print(type(sampleBFiltered))


print('p-valie =', '{0:.3f}'.format(stats.mannwhitneyu(sampleAFiltered, sampleBFiltered)[1]))
print('Относительное различие в конверсии:', '{0:.3f}'.format(sampleBFiltered.mean()/sampleAFiltered.mean()-1))
```

**Выводы** 

1. р-критерий (1,6%) значительно меньше установленного порогового значения в 5%. В результате у нас есть основания отвергнуть нулевую гипотезу и признать, что конверсии групп А и В, скорее всего, различаются.
2. Относительное различие конверсии (у группы В она выше, чем у группы А) на уровне 17,4% является значимым для бизнеса.
3. Этот расчет сделан на очищенных данных. Он является более показательным для принятия решений.
4. Неочищенные данные приводили к тем же результатам. Таким образом оба расчета свидетельствуют об одном и том же: конверсия в группе В выше, чем конверсия в группе А. Разница составляет около 17%.

### Статистическая значимость различий в среднем чеке заказа между группами по «очищенным» данным
