## Предобработка данных

### Начальная обработка

В названиях колонок достаточно одной правки (и та стилистическая)

```python
data = data.rename(columns = {'cityCenters_nearest':'city_centers_nearest'})
```

Большинство данных числовые (некоторые булевы). Но сейчас важно, что колонка с названиями населенных пунктов текстовая. Там могут быть условия для сокрытия неявных дубликатов.

```python
print(data['locality_name'].sort_values().unique())
```

Видны проблемы: "поселок - посёлок", "посе(ё)лок станции - посе(ё)лок при железнодорожной станции", "городской поселок - поселок городского типа" и др.

Можно было бы воообще убрать административные части названий, но с ними информация для анализа будет полнее (нагляднее), особенно с учетом того, что города имеют свой отличительный признак как бы по умолчанию - отсутствие административного статуса. Поэтому ограничимся приведением к единообразному виду.

```python
data['locality_name'] = data['locality_name'].str.replace('посёлок', 'поселок')
data['locality_name'] = data['locality_name'].str.replace("городской поселок", "поселок городского типа")
data['locality_name'] = data['locality_name'].str.replace("поселок при железнодорожной станции", "поселок станции")
data['locality_name'] = data['locality_name'].str.replace('садоводческое некоммерческое товарищество', 'СНТ')
data['locality_name'] = data['locality_name'].str.replace('садовое товарищество', 'СНТ')
# print(data['locality_name'].sort_values().unique())
```

### Обработка пропусков

Пойдем последовательно - слева направо.

**celling_heit** — высота потолков (м) - отсутствует примерно у 40% данных - слишком много для удаления соответствующих строк. Предполагаем, что пропуски связаны с тем, что продавцы не указывали ее в случае, когда этот параметр не повышал продажную стоимость квартиры, не был привлекательным. Можно было бы заменить пропуски медианным значением, но в силу указанного обстоятельства медианное значение дало бы завышенный результат, так как имеющийся набор заполненных данных по высоте потолков имеет несколько "рекламный" характер. Но сначала посмотрим на данные по высоте.

```python
data['ceiling_height'].describe()
```

Оказывается, минимум равен 1 метру! А есть и 100-метровые! Явно недостверные данные. Посмотрим на структуру потолков по высоте.

```python
data['ceiling_height'].value_counts(bins=30)
```

Еще веселей! Оказывается, потлоки от 5 до 30 и даже 100 метров не единичны! Но с ними разберемся потом, в аналитической части проекта (если они на что-то повлияют). А пока примем решение заполнить пропуски первым квартилем (2,52 м), так как медиана в данных обстоятельствах имеет завышенный результат (именно невысокие потолки не указываются продавцами). Для наглядности примем ровно 2,5 метра для заполнения пропусков (это, по сути, стандарт массовой застройки на протяжении десятилетий).

```python
data['ceiling_height'] = data['ceiling_height'].fillna(2.5)
data['ceiling_height'].describe()
```

Теперь и обновленная медиана опустилась до 2,5 м - нормально.

**floors_total** — число этажей в доме. Это не самый существенный параметр (и пропусков пренебрежимо мало). Важнее этаж квартиры. Поэтому заменим пропуски этажности дома исходя из того, что продаваемая квартира находится на верхнем этаже. Причиной пропусков, кстати, могла быть психологическая "рациональность" (зачем указывать одно и то же дважды).

Кроме того, сразу напрашивается приведение количества этажей к целочисленному формату. Это соответствует физическому измерению величины, а также сократит потребный объем памяти и время на обработку данных.

```python
data['floors_total'] = data['floors_total'].fillna(data['floor']).astype(int)

print(data['floors_total'])

data.groupby('floors_total').agg({'floors_total':'count'})
```

Пропусков нет (длина колонки с данными достигла 23699, все целочисленные). Есть, правда, сверхвысотные дома до 60 м, которыех нет в Санкт-Петербурге и Ленинградской области. Данные по ним недостоверны, но они единичны. И в силу этого, не требуют правки. Оставляем их как есть.

**living_area** — жилая площадь в квадратных метрах (м²). Важнейший параметр. Пропущен 1903 из 23699, т.е. 8%.

Посмотрим, как эти пропуски связаны с пропусками площади кухни (**kitchen_area**, по количеству они схожи).

```python
data[(data['living_area'].isna() & data['kitchen_area'].isna())].shape
```

Совпадение значительно. Получается, что в 1464 лучаях пропуски в жилой площади (1903) и кухни (2278) совпадают. Это наталкивает на мысль, что, возможно, 1464 случая - это студии, где площадь кухни неотделима от жилой. 

```python
data['studio'].value_counts()
```

Но студий всего 149.

```python
data[(data['living_area'].isna() & data['kitchen_area'].isna() & data['studio'] == True)].shape
```

А пересечений пропусков площадей кухонь и жилых со студиями и того меньше (всего 10). То есть причина не в "студийности".

Поэтому для целей исследования будет достаточно, если пропуски в жилой площади (и площади кухни) заменим в пропорции их медианных долей.

```python
percentage_of_living_area = data['living_area'] / data['total_area']
percentage_of_living_area_median = percentage_of_living_area.median()

percentage_of_kitchen_area = data['kitchen_area'] / data['total_area']
percentage_of_kitchen_area_median = percentage_of_kitchen_area.median()

print(percentage_of_living_area_median, percentage_of_kitchen_area_median)
```

Коэффицианты выглядят правдоподобно. Применим.

```python
data['living_area'] = data['living_area'].fillna(percentage_of_living_area_median * data['total_area'])
data['kitchen_area'] = data['kitchen_area'].fillna(percentage_of_kitchen_area_median * data['total_area'])

# print(data['living_area'], data['kitchen_area'])
```

Исследуем корреляцию между пропусками в количестве водоемов и парков, с одной стороны, и расстоянием до ближайего такого объекта в 3000 м. Стыкуются ли эти данные? Возможно, так станет более ясной причина пропусков во всех четырех колонка.

```python
data['parks_around3000'].isna().sum() # Пропуски по количеству парков в радиусе 3 км
```

Тогда проверим, нет ли пропусков данных по количеству парков, если указано расстояние до парка менее 3 км.

```python
data[(data['parks_nearest'] < 3000 & ~data['parks_around3000'].isna())].shape
```

Нет таких. Тогда спокойно заполняем пропуски по количеству парков нулями (с одновременным интованием значений в этой колонке), а пропуски по расстоянию до парка условной величиной 10000 м.

```python
data['parks_nearest'] = data['parks_nearest'].fillna(10000)
```

### Удаление редких и выбивающихся значений

**Площадь**

Создадим функцию для стандартной обработки.

```python
# На вход фукции подается объект Series.
# На выходе - описание статистических данных и физические границы "усов"
# (по полтора интерквартильного интервала по обе стороны "ящика").
def margins_of_cutting_off(data_series):
    stats = data_series.describe()
    print(stats)

    q25 = stats.iloc[4]
    q75 = stats.iloc[6]

    left_margin = 2.5 * q25 - 1.5 * q75
    right_margin = 2.5 * q75 - 1.5 * q25
    print(f'\nНа дистанции 1,5 интерквартильного интервала:\
    \n- граница отсечки слева: {left_margin:.2f}\n- граница отсечки справа: {right_margin:.2f}')
```

```python
margins_of_cutting_off(data['total_area'])
```

Левая граница отрицательна - ничего не осекаем слева. Расчетная граница справа выглядит короткой. На основе диаграмм, построенных выше, для дальнейшего поиска зависимостей примем правую границу на уровне 150.
